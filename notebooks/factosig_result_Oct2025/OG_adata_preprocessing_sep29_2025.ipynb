```python
# Exploration of CD34pos.annotation.csv and CD34pos.annotation.w.score.csv

print("--- Loading Annotation Files ---")
cd34_anno_df = pd.read_csv(os.path.join(data_dir, 'CD34pos.annotation.csv'), index_col=0)
cd34_score_df = pd.read_csv(os.path.join(data_dir, 'CD34pos.annotation.w.score.csv'), index_col=0)

# --- Analysis of cd34_anno_df ---
print("\n\n--- Analysis of CD34pos.annotation.csv ---")
print(f"Shape: {cd34_anno_df.shape}")

# Compare indices with og_adata
cd34_anno_barcodes = set(cd34_anno_df.index)
og_adata_barcodes = set(og_adata.obs.index)
common_barcodes_anno = og_adata_barcodes.intersection(cd34_anno_barcodes)
print(f"Overlapping cells with og_adata: {len(common_barcodes_anno)}")

# Analyze columns
print(f"Columns: {cd34_anno_df.columns.tolist()}")
print("\nValue counts for 'predicted.annotation':")
print(cd34_anno_df['predicted.annotation'].value_counts(dropna=False))


# --- Analysis of cd34_score_df ---
print("\n\n--- Analysis of CD34pos.annotation.w.score.csv ---")
print(f"Shape: {cd34_score_df.shape}")

# Compare indices with og_adata
cd34_score_barcodes = set(cd34_score_df.index)
common_barcodes_score = og_adata_barcodes.intersection(cd34_score_barcodes)
print(f"Overlapping cells with og_adata: {len(common_barcodes_score)}")

# Analyze columns
print(f"Columns (cell types with scores): {cd34_score_df.columns.tolist()}")


# --- Cross-Comparison ---
print("\n\n--- Comparing the two annotation files ---")
if cd34_anno_df.index.equals(cd34_score_df.index):
    print("The two files have the exact same cell barcodes in the same order.")
else:
    print("The cell barcodes in the two files are NOT identical.")
```

```python
# Integrate selected columns from CD34pos.annotation.csv into og_adata

# Columns to integrate
columns_to_add = [
    'predicted.annotation',
    'predicted.annotation.score',
    'predicted.pseudotime',
    'predicted.pseudotime.score'
]

# Loop through the columns and add them to og_adata.obs
for col in columns_to_add:
    # Use .reindex() to ensure the data is aligned correctly by cell barcode
    # It will automatically fill with NaN for any cells in adata not in the csv (though we know they all are)
    og_adata.obs[col] = cd34_anno_df[col].reindex(og_adata.obs.index)

print("--- Integration Complete ---")
print(f"Added the following columns to og_adata.obs: {columns_to_add}")

# Verification
print("\n--- Verification ---")
# Check the first few rows of the newly added columns
print(og_adata.obs[columns_to_add].head())

# Check for any NaN values to be sure the merge was successful
print(f"\nNumber of NaN values in 'predicted.annotation' after merge: {og_adata.obs['predicted.annotation'].isna().sum()}")
```

```python
# 5. Check consistency of common columns for a sample of cells
print("\n--- Consistency Check for Common Columns (sample) ---")
sample_barcodes = list(common_barcodes)[:5] # Check first 5 common cells
comparison_df = og_adata.obs.loc[sample_barcodes, list(common_cols)].join(cn_label_df.loc[sample_barcodes, list(common_cols)], lsuffix='_adata', rsuffix='_csv')
```

```python
# Programmatic Sanity Check for common columns (Corrected for Categorical types)

all_consistent = True
# Loop through each of the original common column names
for col in common_cols:
    # Define the column names with suffixes
    adata_col = f"{col}_adata"
    csv_col = f"{col}_csv"
    
    # Convert both columns to string type for a robust comparison.
    # This handles different dtypes (like Categorical) and NaNs gracefully.
    # NaNs will be converted to the string 'nan' in both series.
    series_adata = comparison_df[adata_col].astype(str)
    series_csv = comparison_df[csv_col].astype(str)
    
    if not series_adata.equals(series_csv):
        print(f"Inconsistency found in column: '{col}'")
        all_consistent = False
        # Optional: show the rows with inconsistencies
        # print(comparison_df[~series_adata.eq(series_csv)])

if all_consistent:
    print("Sanity check passed: All common columns are consistent for the sampled cells.")
else:
    print("Sanity check failed: Inconsistencies were found.")
```

```python
# Annotate CN.label based on the logic from 13_all.CN.label.recipient.csv

# Create a series from the csv for easy mapping
cn_label_series = cn_label_df['CN.label']

# Initialize the new column in og_adata.obs. We can use 'unassigned' as a default.
og_adata.obs['CN.label'] = 'unassigned'

# 1. Annotate donor cells as 'normal'
# From our exploration, 'source' column in og_adata defines donor vs. recipient
donor_mask = og_adata.obs['source'] == 'donor'
og_adata.obs.loc[donor_mask, 'CN.label'] = 'normal'
print(f"Annotated {donor_mask.sum()} donor cells as 'normal'.")

# 2. Annotate recipient cells using the cn_label_series
# First, identify the recipient cells in og_adata that have a label in the CSV
recipient_cells_in_adata = og_adata.obs.index[og_adata.obs['source'] == 'recipient']
barcodes_to_update = cn_label_series.index.intersection(recipient_cells_in_adata)

# Use .loc to update the values for these specific cells
og_adata.obs.loc[barcodes_to_update, 'CN.label'] = cn_label_series.loc[barcodes_to_update]
print(f"Annotated {len(barcodes_to_update)} recipient cells from cn_label_df.")


# 3. Verification
print("\n--- Verification of new 'CN.label' column ---")
print("Value counts for 'CN.label' in og_adata.obs:")
print(og_adata.obs['CN.label'].value_counts(dropna=False))

# Check if any recipient cells remain 'unassigned'
unassigned_recipients = og_adata.obs[(og_adata.obs['source'] == 'recipient') & (og_adata.obs['CN.label'] == 'unassigned')]
print(f"\nNumber of recipient cells that remain 'unassigned': {len(unassigned_recipients)}")
if len(unassigned_recipients) > 0:
    print("This is expected for recipient cells present in og_adata but not in cn_label_df.")
    num_recipient_cells = len(og_adata.obs[og_adata.obs['source'] == 'recipient'])
    num_recipient_cells_with_label = len(og_adata.obs[(og_adata.obs['source'] == 'recipient') & (og_adata.obs['CN.label'] != 'unassigned')])
    print(f'Number of recipient cells with label: {num_recipient_cells_with_label} out of {num_recipient_cells}. ({num_recipient_cells} - {num_recipient_cells_with_label} = {num_recipient_cells - num_recipient_cells_with_label})')
    print(f'Percentage: {num_recipient_cells_with_label/num_recipient_cells}')
```

```python
# Integrate timepoint data from sample_collection_time.csv (Corrected)

# 1. Create a "lookup map" from the sample_time_df.
# We set 'sample' as the index, so we can look up values from other columns.
sct_map = sample_time_df.set_index('sample')['Days from SCT']
relapse_map = sample_time_df.set_index('sample')['Days.from.Relapse']

# 2. Use the .map() function to create the new columns in og_adata.obs
# This will look at each value in og_adata.obs['sample'] and find the corresponding
# value in our maps.
og_adata.obs['Days from SCT'] = og_adata.obs['sample'].map(sct_map)
og_adata.obs['Days.from.Relapse'] = og_adata.obs['sample'].map(relapse_map)


print("--- Integration Complete ---")
print("Added 'Days from SCT' and 'Days.from.Relapse' to og_adata.obs.")

# 3. Verification (this part remains the same)
print("\n--- Verification ---")
new_cols = ['Days from SCT', 'Days.from.Relapse']
print(og_adata.obs[new_cols].head())

# Check how many cells got the new data
sct_nan_count = og_adata.obs['Days from SCT'].isna().sum()
relapse_nan_count = og_adata.obs['Days.from.Relapse'].isna().sum()

print(f"\nNumber of cells without 'Days from SCT' data (NaNs): {sct_nan_count}")
print(f"Number of cells without 'Days.from.Relapse' data (NaNs): {relapse_nan_count}")
print(f"This should correspond to cells where the 'sample' value was not in the lookup CSV (e.g., 'unknown_unknown').")
```
